```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
```

```python
# Load train and test data
train_df = pd.read_csv('../data/train.csv')

print("Train shape:", train_df.shape)
```

```python
display(train_df.head())
print(train_df.isnull().sum().to_string())
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x=train_df['label'])
plt.title('Class Distribution of Digits'); plt.xlabel('Digit'); plt.ylabel('Count'); plt.tight_layout()
plt.show()
```

```python
def show_grid(df, n=12):
    idx = np.random.choice(len(df), n, replace=False)
    imgs = df.iloc[idx, 1:].values.reshape(n, 28, 28)
    labels = df.iloc[idx, 0].values
    cols = 6; rows = n//cols + int(n%cols>0)
    plt.figure(figsize=(12, 5))
    for i,(im,lbl) in enumerate(zip(imgs, labels)):
        ax = plt.subplot(rows, cols, i+1)
        ax.imshow(im, cmap='gray'); ax.set_title(lbl, fontsize=9)
        ax.axis('off')
    plt.tight_layout(); plt.show()

show_grid(train_df, n=12)
```
